Survey Participation finder
This is a trial file and the corresponding data CSV files are present in the same folder

```python
import pandas as pd
survey_list = pd.read_csv('data_sheet.csv',header=None,names=['empnames'])
full_emp_list = pd.read_csv('emp_list.csv',header=None,names=['empnames'])
```

```python
survey_list = survey_list.applymap(lambda x: x.lower())
survey_list = survey_list.drop_duplicates()
survey_list
```

```python
full_emp_list = full_emp_list.applymap(lambda y : y.lower())
```

```python
full_emp_list = full_emp_list.drop_duplicates()
full_emp_list
```

```python
tooksurveylist = full_emp_list.assign(tookSurvey=full_emp_list['empnames'].isin(survey_list['empnames']))
tooksurveylist[tooksurveylist['tookSurvey'] == True].count()
```

```python
count = 0
for emp_name in full_emp_list['empnames']:
    print(f"Checking for {emp_name}\n")
    if (survey_list['empnames'].str.contains(emp_name)).any():
        print(f"{emp_name} took the survey")
    else:
        print(f"{emp_name} did not take the survey")
        count += 1
print(f"Total No who did not take the survey: {count}")
```

```python
count = 0

for emp_name in full_emp_list['empnames']:
    found = False
    first_name = emp_name.split()[0]
    for sur_emp_name in survey_list['empnames']:
        if(sur_emp_name.find(first_name) >= 0):
            count += 1
            found = True
    if found == False:
        print(f"{emp_name} missed taking the survey!")
    
print(f"Total Missed: {full_emp_list['empnames'].count() - count}")
```
